 Q2. Explain the concept of function arguments and parameters in Python.

      >>Parameter:-It is the input variables that are defined when definina function.It serve as placeholders for the actual values that will be passed to the function when it is called.
      example
      
       def area(length, width):
      # Parameters: length, width (Placeholder Values)
          return length * width
      
      In this example, length and width are the parameters of the area() function. When this function is called, the value that is passed to length and width will be used to generate the value for the return statement.

      Arguments:-It is the actual values that are passed to a function when it is called. In other words, they are the values that are used to replace the placeholders (parameters) in the function definition.
      example
      
       def area(length, width):
       # Parameters: length, width (Placeholder Values)
           return length * width    
       
       # Arguments: 10, 20 (Actual values)
       result = area(10, 20)
       print(result)

       In this example, 10, and 20 are the arguments that are passed to the area() function. When the function is called, 10 and 20 are used to replace the length and width parameters in the function definition. As a result, the function returns 10 * 20 = 200 to the console.

Q6.  Explain the concept of generators in Python and how they are defined?
     
     >>A generator is a function that produces a sequence of values lazily, one at a time, rather than storing all values in memory at once. It allows for efficient iteration over large datasets or infinite sequences without loading the entire dataset into memory.
     Generator are implemented using the yield statement in Python
     
     def square_no_generator(n):
         for i in range(n):
         yield i**2
     When a generator function is called, it returns a generator object, which behaves like an iterator    


Q7.  What are the advantages of using generators over regular functions?

    >>Here is the advatanges of using generators over regular functions
    1. Lazy Evaluation:
     Generators and iterators use lazy evaluation, meaning they generate values on-the-fly and don’t store all values in memory at once. This is particularly useful when dealing with large datasets or infinite sequences.
     2. Memory Efficiency:
     Regular function can consume a lot of memory if they generate a large list. Generators, on the other hand, generate values one at a time, conserving memory.
    3. Code Simplicity:
     Generators provide a more concise and readable way to create iterators. They eliminate the need to manage the state of the iteration manually, making the code cleaner.
    4. Infinite Sequences:
     Generators allow you to represent infinite sequences efficiently, as they generate values as needed without the need to precompute and store the entire sequence.

     numbers = [1, 2, 3, 4]
     doubled = []
     for num in numbers:
         doubled.append(num * 2)

     print(doubled)  # Output: [2, 4, 6, 8]

     Using map() (short way)
     numbers = [1, 2, 3, 4]
      doubled = list(map(lambda x: x * 2, numbers))

      print(doubled)  # Output: [2, 4, 6, 8]

      When to Use map()?
      When you need to apply a function to every item in a list.
      When you want cleaner and shorter code instead of a for loop.

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?


      >>
      1. map(): Applies a Function to Every Item
       Think of map() like a machine that modifies every item in a list. It takes each item, applies a function, and gives back a new list with the changed values.
       
       numbers = [1, 2, 3, 4]
       doubled = list(map(lambda x: x * 2, numbers))

       print(doubled)  # Output: [2, 4, 6, 8]

      2. filter(): Keeps Only Certain Items
       filter() is like a sieve—it goes through a list and keeps only the items that match a condition, throwing away the rest.

       numbers = [1, 2, 3, 4, 5, 6]
       evens = list(filter(lambda x: x % 2 == 0, numbers))

       print(evens)  # Output: [2, 4, 6]

     3. reduce(): Combines Everything into One Value
      reduce() is like a pressing machine—it combines all items into a single result by applying a function repeatedly.

      from functools import reduce

       numbers = [1, 2, 3, 4]
       product = reduce(lambda x, y: x * y, numbers)

       print(product)  # Output: 24  (1 * 2 * 3 * 4)

```python
"""
Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
"""
def even_sum(numbers):
    sum=0
    for i in numbers:
        if i%2==0:
            sum=sum+i
    return sum

even_sum([1,2,3,4,5,56,8,9])
```

```python
"""
Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.
"""

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
"""
Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
"""
def powers_of_two(n):
    for i in range(n):
        yield 2**i

for power in powers_of_two(5):
    print(power)
```

```python
"""
Q7.Implement a generator function that reads a file line by line and yields each line as a string.
"""
def read_file_lines(filepath):
  try:
    with open(filepath, 'r') as file:
      for line in file:
        yield line
  except FileNotFoundError:
    print(f"Error: File not found at {filepath}")
    return  # Stop iteration if file not found

filepath = 'my_file.txt'  # Replace with your file path

with open(filepath, 'w') as f:
  f.write("This is the first line.\n")
  f.write("This is the second line.\n")
  f.write("This is the third line.\n")

for line in read_file_lines(filepath):
  print(f"Line: {line.strip()}")
```

```python
"""
Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
"""
cel_to_fahren = lambda c: (c * 9/5) + 32

cel_temps = [0, 27, 33, 37, 46, 70]

fahren_temps = list(map(cel_to_fahren, cel_temps))

print(f"Temperatures in Fahrenheit: {fahren_temps}")
```

```python
"""
Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
"""

def remove_vowels(s):
    return "".join(filter(lambda char: char.lower() not in "aeiou", s))

print(remove_vowels("Hello, World!"))
```
